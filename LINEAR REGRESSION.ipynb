```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  
from sklearn.linear_model import LinearRegression
```

```python
data= pd.read_csv('1.02.+Multiple+linear+regression.csv')
data
```

```python
data.head()
```

```python
data.describe()
```

###DECLARE DEPENDENT AND INDEPENDENT VARIABLES

```python
x=data[['SAT','Rand 1,2,3']]
y=data['GPA']
```

###REGRESSION ITSELF

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
##CHECK FOR THEIR R^2
```

```python
reg.score(x,y)
```

```python
#coef 
reg.coef_
```

```python
#reg.intercept
reg.intercept_
```

# ADJUSTED R^2 n=84 (number of observations)
# p=2 (number of predictors)

```python
x.shape
```

```python
# Calculate adjusted R-squared
n = len(y) # or n = x.shape[0]
p = x.shape[1]
r2=reg.score(x,y)
adjusted_r2 = 1 - (1-r2)*(n-1)/(n-p-1)

print("Adjusted R-squared:", adjusted_r2)
```

### adjusted_r2 is <r2 so one or more of the predictors have little or no explanatory power

```python
reg.predict([[1223,1]])
```

```python
# After displaying y_test, we find what the issue is
# The old indexes where preserved (recall earlier in that code we made a note on that)
# The code was: data_cleaned = data_4.reset_index(drop=True)
# Therefore, to get a proper result, we must reset the index and drop the old indexing
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
# Additionally, we can calculate the difference between the targets and the predictions\n",
# Note that this is actually the residual (we already plotted the residuals)\n",
df_pf['Residual'] = df_pf['Target'] - df_pf['Prediction']
# Since OLS is basically an algorithm which minimizes the total sum of squared errors (residuals) and is achieved through minimizing the sum of squared
#residuals. therefore examining the residual is the same as examining the heart of the optimization algorithm
# this comparison makes a lot of sense"
```

```python
df_pf
```

```python
#Finally, it makes sense to see how far off we are from the result percentage-wise\n",
# Here, we take the absolute difference in %, so we can easily order the output and targets
df_pf['Difference%'] = (df_pf['Residual']/df_pf['Target']*100)
df_pf
```

```python
## TO DISPLAY THE ABSOLUTE  DIFF IN PERCENTAGE WE ADD -- np.absolute
df_pf['Difference%'] = np.absolute(df_pf['Residual']/df_pf['Target']*100)
df_pf
```
