```python
import pandas as pd
import numpy as np
```

```python
schools= pd.read_csv('schools.csv')
schools
```

```python
schools.shape
```

```python
schools.describe()
```

```python
#Â Which schools are best for math?
best_math_schools = schools[schools["average_math"] >= 640][["school_name", "average_math"]].sort_values("average_math", ascending=False)
best_math_schools
```

```python
# Calculate total_SAT per school
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
schools
```

```python
# top 10 schools by total_SAT
top_10_schools = schools.sort_values("total_SAT", ascending=False)[["school_name", "total_SAT"]].head(10)
top_10_schools
```

```python
# OR
# top 10 performing schools
top_10_schools = schools[["school_name", "total_SAT"]].sort_values("total_SAT", ascending=False).head(10)
top_10_schools
```

```python
# NYC borough has the highest standard deviation for total_SAT?
boroughs=schools.groupby('borough')['total_SAT'].agg(['count','mean','std']).round(2)
boroughs
```

```python
# find the borough with max std
large_std_dev= boroughs[boroughs["std"] == boroughs["std"].max()]
large_std_dev
```

```python
# Renaming the columns for clarity 
# inplace=True: This argument modifies the DataFrame directly. If you omit inplace=True or set it to False (which is the default), 
# rename() will return a new DataFrame with the columns renamed, leaving the original DataFrame unchanged.

largest_std_dev = large_std_dev.rename(columns={"count": "num_schools", "mean": "average_SAT", "std": "std_SAT"})
largest_std_dev
```
